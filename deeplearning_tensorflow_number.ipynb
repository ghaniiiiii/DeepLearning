```
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

# untuk mengambil gambar gambar dari library yang berukuran 28x28

```
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```
len(X_train)
```

```
len(X_test)
```

```
plt.matshow(X_train[0])
```

```
X_train[0].shape
```

```
X_train = X_train /255
X_test = X_test /255
```

```
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```
X_train_flattened.shape
```

```
X_train_flattened[0]
```

# proses membentuk ann

```
# untuk membuat layer ann
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation="sigmoid")
])

#
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=['accuracy']
)

model.fit(X_train_flattened,y_train,epochs=10)
```

 # evaluasi model

```
model.evaluate(X_test_flattened,y_test)
```

#membuat prediksi

```
y_pred = model.predict(X_test_flattened)
y_pred[1]
```

```
np.argmax(y_pred[1])
```

```
y_pred_labels=[np.argmax(i) for i in y_pred]
y_pred_labels[:5]
```

```
y_test[:5]
```

# membuat confusion matriks

```
cm =tf.math.confusion_matrix(labels = y_test, predictions = y_pred_labels)
cm
```

# membuat tampilanya agar lebih mudah di pahami

```
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm,annot=True,fmt="d")
plt.xlabel("predicted")
plt.ylabel("truth")
```

#mencoba menambahkan layer

```
# untuk membuat layer ann
model = keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation="relu"),
    keras.layers.Dense(10,activation="sigmoid")
])

#
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=['accuracy']
)

model.fit(X_train_flattened,y_train,epochs=10)
```

```
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm,annot=True,fmt="d")
plt.xlabel("predicted")
plt.ylabel("truth")
```

## alternatif agar tidak memakai flattened

```
# untuk membuat layer ann
model = keras.Sequential([
    keras.layers.flatten(input_shape=(28,28))
    keras.layers.Dense(100,input_shape=(784,),activation="relu"),
    keras.layers.Dense(10,activation="sigmoid")
])

#
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=['accuracy']
)

model.fit(X_train,y_train,epochs=10)
```
